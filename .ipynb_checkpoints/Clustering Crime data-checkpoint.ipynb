# Importing all the necessary libraries

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
crime_data = pd.read_csv("crime_data (1).csv")
crime_data
```

```python
crime_data=crime_data.drop(labels=['Unnamed: 0'],axis=1)
crime_data
```

### Normalization

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
crime_data = norm_func(crime_data)
crime_data
```

### Dendogram with all three methods

```python
dendrogram = sch.dendrogram(sch.linkage(crime_data, method='average'))
```

```python
dendrogram = sch.dendrogram(sch.linkage(crime_data, method='single'))
```

```python
dendrogram = sch.dendrogram(sch.linkage(crime_data, method='complete'))
```

### Hierarchical Clustering

```python
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'single')
hc
```

```python
crime_data_Agglomerative=hc.fit_predict(crime_data)
crime_data_Agglomerative
```

```python
Clusters_Agglomerative=pd.DataFrame(crime_data_Agglomerative,columns=['Clusters'])
Clusters_Agglomerative
```

```python
crime_data['H_clusters'] = Clusters_Agglomerative
```

```python
crime_data
```

#### Hierarchical Clustering with cluster 0

```python
crime_data[crime_data['H_clusters']==0]
```

#### Hierarchical Clustering with cluster 1

```python
crime_data[crime_data['H_clusters']==1]
```

#### Hierarchical Clustering with cluster 2

```python
crime_data[crime_data['H_clusters']==2]
```

#### Hierarchical Clustering with cluster 3

```python
crime_data[crime_data['H_clusters']==3]
```

# K Means Clustering

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(crime_data)
    wcss.append(kmeans.inertia_) #wcss
   
```

```python
wcss
```

#### Finding optimized n value using Elbow Method

```python
plt.plot(range(1,10),wcss)
```

```python
x=KMeans(n_clusters=3,max_iter=500,algorithm='auto')
```

```python
x.fit(crime_data)
```

```python
clusters=x.predict(crime_data)
```

```python
crime_data['K_clusters']=clusters
```

```python
crime_data
```

```python
crime_data=crime_data.drop(labels=["clusters"],axis=1)
```

```python
crime_data[crime_data['K_clusters']==0]
```

```python
crime_data[crime_data['K_clusters']==1]
```

```python
crime_data[crime_data['K_clusters']==2]
```

```python
crime_data[crime_data['K_clusters']==3]
```

##### Model is more optimized in k-means to hireachical, in kmeans have two clusters and in hierachical only one

# DBSCAN

```python
dbs_cluster=dbs.fit_predict(crime_data)
```
